# 🎙️ Atomic Step 1: Audio Ingestion & 16kHz Mono Standardization

This notebook resamples and standardizes raw audio files from Google Drive to **16kHz mono WAV** format. Standardizing the audio is a critical pre-requisite for running Pyannote speaker diarization and Whisper models.

```python
from google.colab import drive
import os

# Mount Google Drive
try:
    drive.mount('/content/drive')
    print("[SUCCESS] Google Drive mounted.")
except Exception:
    print("[INFO] Already mounted or skipped.")
```

## ⚙️ Parameters

```python
# @markdown ### 📁 Audio Paths
input_audio_path = "/content/drive/MyDrive/AnnamAI Tasks/Outreach Activity STT + Question Generation Workflow/Atomic Notebooks/Sample Audio Files/MarauliKhurad1.m4a" # @param {type:"string"}
standardized_audio_folder = "/content/drive/MyDrive/AnnamAI Tasks/Outreach Activity STT + Question Generation Workflow/Atomic Notebooks/Standardized_Audio/" # @param {type:"string"}

if not os.path.exists(input_audio_path):
    print(f"[ERROR] Input audio not found at: '{input_audio_path}'")
else:
    audio_filename = os.path.basename(input_audio_path)
    audio_name_only, _ = os.path.splitext(audio_filename)
    os.makedirs(standardized_audio_folder, exist_ok=True)
    print(f"[SUCCESS] Validated input path. Output will be saved in: {standardized_audio_folder}")
```

## 🎵 Process Audio

```python
import librosa
import soundfile as sf

if os.path.exists(input_audio_path):
    print(f"Ingesting and resampling audio: '{input_audio_path}'")
    # Load and resample to 16kHz mono in-memory
    y, sr = librosa.load(input_audio_path, sr=16000, mono=True)
    print(f"[SUCCESS] Loaded audio. Sample Rate: {sr} Hz, Samples Count: {len(y)}")
    
    # Export standardized WAV file
    output_wav_path = os.path.join(standardized_audio_folder, f"{audio_name_only}_standardized.wav")
    sf.write(output_wav_path, y, sr)
    print(f"\n[SUCCESS] Standardized audio saved to: '{output_wav_path}'")
```
